```python
sheet = cv2.imread('assets\\images\\1.png')
note = cv2.imread('ss1.png')
w, h = note.shape[:-1]

res = cv2.matchTemplate(sheet, note, cv2.TM_CCOEFF_NORMED)
threshold = 0.70
loc = np.where(res >= threshold)
for pt in zip(*loc[::-1]):  # Switch columns and rows
    cv2.rectangle(sheet, (pt[0]-2, pt[1]-2), (pt[0] + h+1, pt[1] + w+1), (0, 0, 255), 1)

cv2.imwrite('result.png', sheet)
```
